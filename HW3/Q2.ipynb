```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
img_list = []
img_list.append(cv2.imread("images/Q2/1.jpg"))
img_list.append(cv2.imread("images/Q2/2.jpg"))
img_list.append(cv2.imread("images/Q2/3.jpg"))
img_list.append(cv2.imread("images/Q2/4.jpg"))
img_list.append(cv2.imread("images/Q2/5.jpg"))
img_list.append(cv2.imread("images/Q2/6.jpg"))
img_list.append(cv2.imread("images/Q2/7.jpg"))
img_list.append(cv2.imread("images/Q2/8.jpg"))
for i in range(len(img_list)):
  img_list[i] = cv2.cvtColor(img_list[i], cv2.COLOR_BGR2RGB)
  img_list[i]=cv2.resize(img_list[i],(0,0),fx=0.33,fy=0.4)
fig, axs = plt.subplots(1, 8, figsize=(50,4))
for i in range(8):
  axs[i].imshow(img_list[i])
  axs[i].axis('off')
plt.show()
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
stitchy=cv2.Stitcher.create()
(dummy,output)=stitchy.stitch(img_list)
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
if dummy != cv2.STITCHER_OK:
  # checking if the stitching procedure is successful
  # .stitch() function returns a true value if stitching is 
  # done successfully
  print("stitching ain't successful")
else:
  plt.figure(figsize=(17,12))
  plt.imshow(output)
```
