```python
# Q1 ) ج
new_folder = "Q1/Q1_2/"
os.makedirs("Q1/Q1_2")
from pathlib import Path
folder_dir = 'Q1' 
# iterate over files in
# that directory
images = os.listdir(folder_dir)
gray_images = []
dimensions = []
for image in images:
  if (image.endswith(".jpeg")):
    addr = os.path.join(folder_dir, image)
    img = cv2.imread(addr)
    grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # gray_images.append(grayscale)
    dimension = str(grayscale.shape[0]) + "," + str(grayscale.shape[1])
    cv2.imwrite(os.path.join(new_folder, image), grayscale)
    txt_name = image[:-len(".jpeg")]
    txt_name += ".txt"
    txt_name = os.path.join(new_folder, txt_name)
    f = open(txt_name, "w")
    f.write(dimension)
    f.close()
```
